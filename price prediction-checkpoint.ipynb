```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime
import math
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM, Input

def get_historical_data(quote):
    end = datetime.now()
    start = datetime(end.year-2, end.month, end.day)
    df = yf.download(quote, start=start, end=end)
    return df

def create_and_train_model(crypto_name, dataset):
    # Plot the historical data
    plt.figure(figsize=(16,8))
    plt.title(f'{crypto_name} Price History')
    plt.plot(dataset['Close'])
    plt.xlabel('Date', fontsize=18)
    plt.ylabel('Close Price USD ($)', fontsize=18)
    plt.grid()
    plt.show()

    dataset = dataset.filter(['Close'])
    dataset_values = dataset.values

    training_data_length = math.ceil(len(dataset_values) * 0.8)

    # Scale the data
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(dataset_values)

    # Create the training dataset
    train_data = scaled_data[0:training_data_length, :]
    x_training = []
    y_training = []

    for i in range(60, len(train_data)):
        x_training.append(train_data[i-60:i, 0])
        y_training.append(train_data[i, 0])

    x_training, y_training = np.array(x_training), np.array(y_training)
    x_training = np.reshape(x_training, (x_training.shape[0], x_training.shape[1], 1))

    # Define the LSTM model
    model = Sequential()
    model.add(Input(shape=(x_training.shape[1], 1)))
    model.add(LSTM(50, return_sequences=True))
    model.add(LSTM(50, return_sequences=False))
    model.add(Dense(25))
    model.add(Dense(1))

    # Compile the model
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Train the model
    model.fit(x_training, y_training, batch_size=1, epochs=7)

    # Create the testing dataset
    test_data = scaled_data[training_data_length - 60:, :]
    x_testing = []
    y_testing = dataset_values[training_data_length:, :]

    for i in range(60, len(test_data)):
        x_testing.append(test_data[i-60:i, 0])

    x_testing = np.array(x_testing)
    x_testing = np.reshape(x_testing, (x_testing.shape[0], x_testing.shape[1], 1))

    # Get the model's predicted price values
    predictions = model.predict(x_testing)
    predictions = scaler.inverse_transform(predictions)

    # Calculate RMSE
    rmse = np.sqrt(np.mean((predictions - y_testing) ** 2))

    # Plot the data
    train = dataset[:training_data_length]
    valid = dataset[training_data_length:]
    valid['Predictions'] = predictions

    plt.figure(figsize=(16,8))
    plt.title(f'{crypto_name} Cryptocurrency Price Prediction LSTM Model')
    plt.xlabel('Date', fontsize=20)
    plt.ylabel('Close Price USD $', fontsize=20)
    plt.plot(train['Close'])
    plt.plot(valid[['Close', 'Predictions']])
    plt.legend(['Training', 'Validation', 'Prediction'], loc='upper left', fontsize=18)
    plt.grid()
    plt.show()

    # Predict the next day's price
    past_60_days = dataset[-60:].values
    past_60_days_scaled = scaler.transform(past_60_days)

    X_test = []
    X_test.append(past_60_days_scaled)
    X_test = np.array(X_test)
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

    prediction_price = model.predict(X_test)
    prediction_price = scaler.inverse_transform(prediction_price)
    print(f'The predicted price for the next day for {crypto_name} is: {prediction_price[0][0]}')

# Ethereum
Eth = get_historical_data('ETH-USD')
create_and_train_model('Ethereum (ETH-USD)', Eth)

# Arbitrum
Arb = get_historical_data('ARB-USD')
create_and_train_model('Arbitrum (ARB-USD)', Arb)

# Chainlink
Link = get_historical_data('LINK-USD')
create_and_train_model('Chainlink (LINK-USD)', Link)
```
